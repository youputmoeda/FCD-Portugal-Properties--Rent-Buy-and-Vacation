```python
df = pd.read_csv('portugal_ads_proprieties.csv')
df.head()
```

```python
#Splitting Strings
df['Location'].str.rsplit(',', expand=True, n=1)
```

```python
#Splitting Strings
df[['City', 'District']] = df['Location'].str.rsplit(',', expand=True, n=1)
```

```python
df.head()
```

```python
df = df.drop(columns=['Location'])
df.head()
```

```python
df.District.unique()
```

```python
df.info()
```

```python
df['District'].value_counts()
```

```python
#Clean white spaces before and after
df['District'] = df['District'].str.strip()
df['District'].value_counts()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#Delete unnecessary
df = df.dropna(subset=['Price'])
```

```python
df.isnull().sum()
```

```python
#Fix missing values
df = df.fillna({'Condition': 'Used'})
df = df.fillna({'Bathrooms': 1})
df = df.dropna(subset=['District'])
df.isnull().sum()
```

```python
df.head()
```

```python
df['Bathrooms'].head()
```

```python
# Replacing Island name for district
df['District'].replace(['Ilha da Madeira', 'Ilha de São Miguel', 'Ilha Terceira', 'Ilha do Pico', 'Ilha de Porto Santo',
                        'Ilha do Faial', 'Ilha de Santa Maria', 'Ilha de São Jorge', 'Ilha das Flores', 'Ilha da Graciosa',
                       'Ilha do Corvo'], ['Madeira', 'Açores', 'Açores', 'Açores', 'Madeira', 'Açores', 'Açores', 'Açores', 'Açores', 'Açores',
                        'Açores'], inplace=True)
df.head()
```

```python
df['District'].value_counts()
```

```python
def long(x):
    if x == 'Porto':
        return 41.1579
    elif x == 'Lisboa':
        return 38.7223 
    elif x == 'Setúbal':
        return 38.5254
    elif x == 'Faro':
        return 37.0194
    elif x == 'Braga':
        return 41.5454
    elif x == 'Aveiro':
        return 40.6405
    elif x == 'Leiria':
        return 39.7495
    elif x == 'Coimbra':
        return 40.2033
    elif x == 'Santarém':
        return 39.2367
    elif x == 'Açores':
        return 37.7412
    elif x == 'Madeira':
        return 32.7607
    elif x == 'Viana do Castelo':
        return 41.6918
    elif x == 'Viseu':
        return 40.6588
    elif x == 'Castelo Branco':
        return 39.8196
    elif x == 'Beja':
        return 38.0153
    elif x == 'Vila Real':
        return 41.3010
    elif x == 'Évora':
        return 38.5714
    elif x == 'Portalegre':
        return 39.2967
    elif x == 'Guarda':
        return 40.5308
    elif x == 'Bragança':
        return 41.8061
```

```python
# Adding Longitude
df['Longitude'] = df['District'].apply(long)
df.head()
```

```python
def lat(x):
    if x == 'Porto':
        return 8.6291
    elif x == 'Lisboa':
        return 9.1393 
    elif x == 'Setúbal':
        return 8.8941
    elif x == 'Faro':
        return 7.9304
    elif x == 'Braga':
        return 8.4265
    elif x == 'Aveiro':
        return 8.6538
    elif x == 'Leiria':
        return 8.8077
    elif x == 'Coimbra':
        return 8.4103
    elif x == 'Santarém':
        return 8.6860
    elif x == 'Açores':
        return 25.5672
    elif x == 'Madeira':
        return 16.9595
    elif x == 'Viana do Castelo':
        return 8.8344
    elif x == 'Viseu':
        return 7.9148
    elif x == 'Castelo Branco':
        return 7.4965
    elif x == 'Beja':
        return 7.8627
    elif x == 'Vila Real':
        return 7.7422
    elif x == 'Évora':
        return 7.9135
    elif x == 'Portalegre':
        return 7.4285
    elif x == 'Guarda':
        return 7.2221
    elif x == 'Bragança':
        return 6.7567
```

```python
# Adding Latitude
df['Latitude'] = df['District'].apply(lat)
df.head()
```

```python
# Reorder Columns
new_columns = ['District', 'City', 'Longitude', 'Latitude', 'Area', 'Rooms', 'Bathrooms', 'Price', 'Condition', 'ProprietyType', 'AdsType']
df = df[new_columns]
df.head()
```
