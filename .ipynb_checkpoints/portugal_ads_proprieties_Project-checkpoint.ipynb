<h4> Longitude and Latitude </h4>

```python
def long(x):
    if x == 'Porto':
        return -8.6291
    elif x == 'Lisboa':
        return -9.1393
    elif x == 'Setúbal':
        return -8.8880
    elif x == 'Faro':
        return -7.9300
    elif x == 'Braga':
        return -8.4226
    elif x == 'Aveiro':
        return -8.6520
    elif x == 'Leiria':
        return -8.8077
    elif x == 'Coimbra':
        return -8.4103
    elif x == 'Santarém':
        return -8.6853
    elif x == 'Açores':
        return -25.6750
    elif x == 'Madeira':
        return -16.9595
    elif x == 'Viana do Castelo':
        return -8.8301
    elif x == 'Viseu':
        return -7.9135
    elif x == 'Castelo Branco':
        return -7.4931
    elif x == 'Beja':
        return -7.8635
    elif x == 'Vila Real':
        return -7.7371
    elif x == 'Évora':
        return -7.9097
    elif x == 'Portalegre':
        return -7.4283
    elif x == 'Guarda':
        return -7.2674
    elif x == 'Bragança':
        return -6.7570
```

```python
import folium
from folium.plugins import MarkerCluster

# Criando um mapa de contagem de propriedades por localização
map_properties = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)

marker_cluster = MarkerCluster().add_to(map_properties)

for idx, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=f"Price: {row['Price']}").add_to(marker_cluster)

# Exibir o mapa diretamente no notebook
map_properties
```

```python
import folium
from folium.plugins import MarkerCluster

# Criando um mapa de contagem de propriedades por cidade
map_city_count = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=7)

marker_cluster_city_count = MarkerCluster().add_to(map_city_count)

for city, count in df['City'].value_counts().items():
    latitude = lat(city)
    longitude = long(city)
    
    if latitude is not None and longitude is not None:
        folium.Marker([latitude, longitude], popup=f"{city}: {count} properties").add_to(marker_cluster_city_count)

# Exibir o mapa diretamente no notebook
map_city_count
```

```python
import folium
from folium.plugins import MarkerCluster

# Criando um mapa de média de preços por cidade
map_avg_price = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=7)

marker_cluster_avg_price = MarkerCluster().add_to(map_avg_price)

for city, avg_price in df.groupby('City')['Price'].mean().items():
    folium.Marker([lat(city), long(city)], popup=f"{city}: Avg Price - {avg_price:.2f}").add_to(marker_cluster_avg_price)

# Exibir o mapa diretamente no notebook
map_avg_price
```

```python
import folium
from folium.plugins import HeatMap

# Criando um mapa de heatmap com a distribuição de preços
map_heatmap = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)

# Convertendo os dados para uma lista de listas [latitude, longitude, peso]
heat_data = [[row['Latitude'], row['Longitude'], row['Price']] for idx, row in df.iterrows()]

# Adicionando o HeatMap ao mapa
HeatMap(heat_data).add_to(map_heatmap)

# Exibir o mapa diretamente no notebook
map_heatmap
```
